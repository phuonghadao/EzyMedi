```python
with open('vctk_generate.py', 'w') as f:
    f.write('''
import torch
import os
import sys
sys.path.append('/content/vits')
sys.path.append('/content/vits/monotonic_align')
import numpy as np
from scipy.io.wavfile import write
import IPython.display as ipd

import commons
import utils
from models import SynthesizerTrn
from text.symbols import symbols
from text import text_to_sequence

# Load config
hps_ms = utils.get_hparams_from_file("/content/vits/configs/vctk_base.json")

# Initialize model
net_g_ms = SynthesizerTrn(
    len(symbols),
    hps_ms.data.filter_length // 2 + 1,
    hps_ms.train.segment_size // hps_ms.data.hop_length,
    n_speakers=hps_ms.data.n_speakers,
    **hps_ms.model
).cpu()
_ = net_g_ms.eval()

# Load pretrained weights
_ = utils.load_checkpoint("/content/pretrained_vctk (1).pth", net_g_ms, None)

# Prepare text function
def get_text(text, hps):
    text_norm = text_to_sequence(text, hps.data.text_cleaners)
    if hps.data.add_blank:
        text_norm = commons.intersperse(text_norm, 0)
    text_norm = torch.LongTensor(text_norm)
    return text_norm

# Sentences to synthesize
sentences = [
    "But it is quite plain to me that all the arrangements for my wedding are going to be made by the Snellings.",
    "The quick brown fox jumps over the lazy dog.",
    "Artificial intelligence is transforming technology.",
    "Speaker verification is crucial for secure systems.",
    "Generating fake speech helps in robustness testing.",
    "The sun sets in the west and rises in the east.",
    "She sells seashells by the seashore.",
    "Cybersecurity remains a top priority in the digital age.",
    "Data privacy is fundamental in modern communication.",
    "Deep learning models require vast amounts of data.",
    "Tomorrow's weather forecast predicts heavy rain.",
    "The cat sat on the mat and looked at the window.",
    "OpenAI develops advanced AI technologies.",
    "Speech synthesis has made significant progress recently.",
    "Python is one of the most popular programming languages.",
    "The conference will start at ten o'clock sharp.",
    "Machine learning algorithms improve over time.",
    "A stitch in time saves nine.",
    "Neural networks are inspired by the human brain.",
    "Voice biometrics provide secure authentication solutions."
]

# Speaker ID
sid = torch.LongTensor([4])

# Create output folder
output_folder = '/content/fake_vctk_audio'
os.makedirs(output_folder, exist_ok=True)

# Loop over sentences
for idx, input_text in enumerate(sentences):
    stn_tst = get_text(input_text, hps_ms)
    with torch.no_grad():
        x_tst = stn_tst.unsqueeze(0)
        x_tst_lengths = torch.LongTensor([stn_tst.size(0)])
        audio = net_g_ms.infer(
            x_tst,
            x_tst_lengths,
            sid=sid,
            noise_scale=0.667,
            noise_scale_w=0.8,
            length_scale=1
        )[0][0,0].data.cpu().float().numpy()

    output_path = f'{output_folder}/fake_speech_vctk_{idx:04}.wav'
    write(output_path, hps_ms.data.sampling_rate, (audio * 32768).astype(np.int16))
    print(f"✅ Fake speech generated at {output_path}")
    ipd.display(ipd.Audio(audio, rate=hps_ms.data.sampling_rate))
''')
```
